```python
import  pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from numpy import array, reshape, zeros, append, arange
import math
from math import log, e, pi
import numpy
from sys import argv
```

# Formation energy as  chemical potential vary


 #### Total Energies in eV

```python
E0=-6108318.476328624 # pristine
E1=-6174864.306138340 # addon S
E2=-6041768.483695555 # mono S vacancy
E3=-5975218.839354995 # di S vacancy up&down
E4=-5975218.551673065 # di S vacancy neighboring
E5=-5997075.714803956 # mono Mo vacancy
E_MoSe2=-244332.785924554 # primtitive
ESe8=-532375.657399940   #8 atoms in unitcell
mu_Se8=ESe8/8
EMo=-222473.348568306  # Total energy in eV
mu_Mobcc=EMo/2

```

 ## $$ E^{d}_{f}=E_{defect}(n_{i}+\Delta n_{i})-E_{pristine}(n_{i}) - \sum_{i} \Delta n_{i} \mu_{i} $$
 
## $$  E^{Bulk}_{MoSe_{2}}-2\mu_{Se8} \leq \mu_{Mo} \leq \mu^{Bulk}_{Mo} $$
## $$ \frac{1}{2}(E_{MoSe2}-\mu^{Bulk}_{Mo})\leq \mu_{Se} \leq \mu_{Se8} $$


```python
mu_Se0_i=0.5*(E_MoSe2-mu_Mobcc)
mu_Mo_i=E_MoSe2-2*mu_Se8
mu_Mo_f=mu_Mobcc
mu_Se0_f=mu_Se8
mu_Se0=arange(mu_Se0_i,mu_Se0_f,0.05)
mu_Mo0=arange(mu_Mo_i,mu_Mo_f,0.05)
```

```python
Ef_addSe =[] 
Ef_VSe = []
Ef_VSe2 =[]
Ef_VSe22=[]
Ef_Mo=[]

for i in mu_Se0:
    Ef_addSe.append(E1-E0-i) 
    Ef_VSe.append(E2-E0+i)
    Ef_VSe2.append(E3-E0+2*i) 
    Ef_VSe22.append(E4-E0+2*i)
 
for j in mu_Mo0:
    Ef_Mo.append(E5-E0+j)

plt.plot(mu_Se0-mu_Se0_f,Ef_addSe,'b', label='addon S') 
plt.plot(mu_Se0-mu_Se0_f,Ef_VSe,'r', label='Mono S vacancy') 
plt.plot(mu_Se0-mu_Se0_f,Ef_VSe22, 'g', label='di S vacancy(neig)') 
plt.plot(mu_Se0-mu_Se0_f,Ef_VSe2, 'k', label='di S vacancy(up&down)') 

plt.xlabel('\Delta mu_Se', fontsize=12)
plt.ylabel('Formation energies eV', fontsize=12)
plt.legend(fontsize=6)
plt.yticks(fontsize=10)
plt.tight_layout()


plt.savefig('Efvsmu_Se.png',dpi=1200)

plt.show()
```

```python
plt.plot(mu_Mo0-mu_Mo_f,Ef_Mo, 'yellow', label='Mono Mo vacancy') 
#plt.plot(T, VMo, 'yellow', label='Mono Mo vacany') 
plt.xlabel('\Delta mu_Mo', fontsize=12)
plt.ylabel('Formation energies eV', fontsize=12)
plt.legend(fontsize=6)
plt.yticks(fontsize=10)
plt.tight_layout()
#
plt.savefig('Efvsmu_Mo.png',dpi=1200)
plt.show()  
```

```python
p0=1e+06  # 
k=1.380649e-16
h=6.62607015e-27
hbar=1.054571817e-27
sigma=8
pi=3.14159265359
m=3.1941378e-22
IA=7.10755979e-23  
IB=8.81126702e-23
IC=1.20492497e-22
R=8.31446261815324
convert=29979245800.0*2*pi #cm^-1 to Hz
```

```python
A=[]
for T in range(100,1010,10):
    A = np.append(np.log((((2*pi*m)**(3/2))*((k*T)**(5/2)))/(p0*(h**3))),A)
#print('A= ', A)
```

```python
B=[]
for T in range(100,1010,10):
    I=np.sqrt(IA)*np.sqrt(IB)*np.sqrt(IC)
    B=np.append(np.log(np.sqrt(pi)/sigma)+ np.log((((8*pi*k*T)/(h**2))**(3/2))*I),B)
#print('B= ', B)
```

```python
C=[]
w=[35.720038406727,35.788201571001,65.360254226816,65.389780783909,90.207394594899,90.227279171666,104.19445300601,
106.72662601322,117.61853788420,117.62193636527,213.54597619628,232.04616772547,232.06262935387,259.45646133170,
259.47716334035,261.04584683119,261.05974864895,262.62491469909]
w=np.array(w)*convert
for T in range(100,1010,10):
    temp=-numpy.array([(np.log(1-math.exp(-(hbar*i)/(k*T)))) for i in w])
    C=np.append(np.sum(temp),C)
```

```python
D=[]
#for T in range(100,1010,10):
temp=numpy.array([(hbar*i)/(2) for i in w])
D=np.sum(temp)
#print('D= ', D)
```

```python
k=8.617333262145e-05 ###########
T=arange(100,1010,10)
mu_0=-k*T*(A+B+C)
plt.plot(T,mu_0, 'red', label='For S8 ring') 
plt.xlabel('T in K', fontsize=12)
plt.ylabel('mu_0 eV', fontsize=12)
plt.legend(fontsize=6)
plt.yticks(fontsize=10)
plt.tight_layout()
#plt.savefig('mu_0vsT.png',dpi=1200)
```

```python
E_DFT=-532375.657399940 #ev
p=100 # 1.89e-3 atm 
#for p in arange(p0/1e+12,p0):
#    print(k*T*np.log(p/p0))
```

```python
T=arange(100,1010,10)
mu_S8=mu_0 + k*T*np.log(p/p0)+D + E_DFT
plt.plot(T,mu_S8, 'red', label='For Se8 ring')
plt.xlabel('T in K', fontsize=12)
plt.ylabel('mu eV', fontsize=12)
plt.legend(fontsize=6)
plt.yticks(fontsize=10)
plt.tight_layout()
```

```python
mu_S=mu_S8/8
plt.plot(T,mu_S, 'blue', label='For Se')
plt.xlabel('T in K', fontsize=12)
plt.ylabel('mu eV', fontsize=12)
plt.legend(fontsize=6)
plt.yticks(fontsize=10)
plt.tight_layout()
#plt.savefig('mu_SvsT.png',dpi=1200)
```
