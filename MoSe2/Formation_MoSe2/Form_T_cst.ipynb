```python
import  pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from numpy import array, reshape, zeros, append, arange
import math
from math import log, e, pi
import numpy
from sys import argv
```

# Formation energy as  chemical potential vary


 #### Total Energies in eV

```python
E0=-6108318.476328624 # pristine
E1=-6174864.306138340 # addon S
E2=-6041768.483695555 # mono S vacancy
E3=-5975218.839354995 # di S vacancy up&down
E4=-5975218.551673065 # di S vacancy neighboring
E5=-5997073.703111252 # mono Mo vacancy
E_MoSe2=-244332.785924554 # primtitive


ESe8=-532375.657399940   #8 atoms in unitcell
mu_Se8=ESe8/8
EMo=-222473.348568306  # Total energy in eV
mu_Mobcc=EMo/2

```

 ## $$ E^{d}_{f}=E_{defect}(n_{i}+\Delta n_{i})-E_{pristine}(n_{i}) - \sum_{i} \Delta n_{i} \mu_{i} $$
 
## $$  E^{Bulk}_{MoSe_{2}}-2\mu_{Se8} \leq \mu_{Mo} \leq \mu^{Bulk}_{Mo} $$
## $$ \frac{1}{2}(E_{MoSe2}-\mu^{Bulk}_{Mo})\leq \mu_{Se} \leq \mu_{Se8} $$


```python
mu_Se0_i=0.5*(E_MoSe2-mu_Mobcc)
mu_Mo_i=E_MoSe2-2*mu_Se0_i
#mu_Mo_f=mu_Mobcc
print(mu_Se0_i)
print(mu_Mo_i)
mu_Se0_f=mu_Se8
mu_Mo_f=E_MoSe2-2*mu_Se0_f
mu_Se0=arange(mu_Se0_i,mu_Se0_f,0.05)
mu_Mo0=arange(mu_Mo_f,mu_Mo_i,0.05)
```

```python
Ef_addSe =[] 
Ef_VSe = []
Ef_VSe2 =[]
Ef_VSe22=[]
Ef_Mo=[]

for i in mu_Se0:
    Ef_addSe.append(E1-E0-i) 
    Ef_VSe.append(E2-E0+i)
    Ef_VSe2.append(E3-E0+2*i) 
    Ef_VSe22.append(E4-E0+2*i)
 
for j in mu_Mo0:
    Ef_Mo.append(E5-E0+j)

plt.plot(mu_Se0-mu_Se0_f,Ef_addSe,'b', label='addon Se') 
plt.plot(mu_Se0-mu_Se0_f,Ef_VSe,'r', label='Mono Se vacancy') 
plt.plot(mu_Se0-mu_Se0_f,Ef_VSe22, 'g', label='di Se vacancy(neig)') 
plt.plot(mu_Se0-mu_Se0_f,Ef_VSe2, 'k', label='di Se vacancy(up&down)') 

plt.xlabel(r'$\Delta \mu_{Se}$', fontsize=12)
plt.ylabel('Formation energies eV', fontsize=12)
plt.legend(fontsize=10)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
#

plt.savefig('Ef_muSe.pdf')

plt.show()
```
