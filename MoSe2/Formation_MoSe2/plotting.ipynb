```python
import  pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from numpy import array, reshape, zeros, append, arange
import math
from math import log, e, pi
import numpy
from sys import argv
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
d0=pd.read_csv('high_T.txt', sep='\s+',header=0)
d1=pd.read_csv('high_P.txt', sep='\s+',header=0)

x1=d0['T']
y11=d0['addS']
y12=d0['VS']
y13=d0['VS2']
y14=d0['VS22']
#print(y14)
x2=d1['P']
y21=d1['addS']
y22=d1['VS']
y23=d1['VS2']
y24=d1['VS22']
ax1.plot(x1,y13, 'k', label='di Se vacancy(up&down)') 
ax1.plot(x1, y11, 'b', label='add atom Se')
ax1.plot(x1, y12, 'r', label='Mono Se vacancy')
ax1.plot(x1, y14, 'g', label='di Se vacancy(neig)') 
#ax1.plot(x1, y14, 'k', label=r'di Se vacancy(up$\&$down)')
#ax1.plot(x1, VMo, 'magenta', label='Mono Mo vacany') 
ax1.set_xlabel('Temperature K', fontsize=12)
ax1.set_ylabel('Formation energies eV', fontsize=12)
#ax1.legend(loc='lower right',fontsize=8)
#ax1.set_xlabel(r'$\Delta \mu_{Se}$', fontsize=12)
#ax1.set_ylabel('Formation energies eV', fontsize=12)


plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
ax2 = ax1.twiny()
ax2.plot(x2,y23, 'k--', label='di Se vacancy(up&down)') 
ax2.plot(x2, y21, 'b--', label='add atom Se')
ax2.plot(x2, y22, 'r--', label='Mono Se vacancy')
ax2.plot(x2, y24, 'g--', label='di Se vacancy(neig)') 
#ax2.plot(x2,y21, 'magenta', label='Mono Mo vacancy') 
ax2.set_xlabel(r'Pressure atm', fontsize=12)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
#
#ax1.legend(loc=4)
ax2.legend(loc=1)
```
